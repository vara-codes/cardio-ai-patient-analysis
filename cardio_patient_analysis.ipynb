```python
import pandas as pd
df = pd.read_csv('cardio_patients.csv')
print("="*60)
print(df)
print("="*60)
print(" First 5 rows :\n",df.head(5))
print("="*60)
print(" last 5 rows :\n",df.tail(5))
print("="*60)
print(df.shape)
print("="*60)
print(df.columns)
print("="*60)
print("Dataset information:\n",df.info())
print("="*60)
print("="*60)
print("\n",df.isnull().sum())
print("="*60)
print(df)
print("="*60)
print(df['HeartRate'].fillna(df["HeartRate"].mean(),inplace=True))
print("="*60)
print(df[df.duplicated()])
print("="*60)
print(df['Risk'].unique())
print("="*60)
print(df['Risk'].nunique())
print("="*60)
print(df.groupby('Risk')['PatientID'].count())
print("="*60)
print("Unique patient names\n",df['Name'].unique())
print("="*60)
print("number of Unique patient names\n",df['Name'].nunique())
print("="*60)
print("Unique Ages \n",df['Age'].unique())
print("="*60)
print("number of Unique Ages \n",df['Age'].nunique())
print("="*60)
print("Patients older than 60 \n",df[df["Age"] > 60] )
print("="*60)
print("High Risk patients\n",df[df['Risk'] == "High"])
print("="*60)
print(df[(df['Age'] > 55) & (df['BloodPressure'] > 130)])
print("="*60)
print(df[(df['Risk'] == "High") & (df['HeartRate'] > 80)])
print("="*60)
print("Age Sorting in ascending",df.sort_values(by=['Age']))
print("="*60)
print("Age in decending order",df.sort_values(by='Age',ascending=False))
print("="*60)
print("BloodPressure in decending order\n ",df.sort_values(by=['BloodPressure'],ascending= False))
print("="*60)
df['HealthScore'] = df["HeartRate"] + df["BloodPressure"]
print(df)
print("="*60)
print(df.groupby('Risk')['HeartRate'].mean())
print("="*60)
print(df.groupby('Risk')['HeartRate'].max())
print("="*60)
print(df.groupby('Risk')['HeartRate'].min())
print("="*60)
print("Average BloodPressure is \n",df.groupby('Risk')['BloodPressure'].mean())
print("="*60)
print("Maximum BloodPressure is \n",df.groupby('Risk')['BloodPressure'].max())
print("="*60)
print("Minimum BloodPressure is \n",df.groupby('Risk')['BloodPressure'].min())
print("="*60)
print("Number of patient in each risk category \n",df.groupby('Risk').value_counts())
print("="*60)
print(df.groupby('Risk').agg({
  "Age" : "mean",
  "HeartRate":"mean",
  "BloodPressure":"max"
}))
print("="*60)
print(df.groupby('Risk')['HealthScore'].agg(["mean","max","min"]))
print("="*60)
print("the highest number of patients are:\n",df.groupby('Risk').value_counts().max())
print("="*60)
print("The patient appear more than one is \n",df['Name'].value_counts())
print("="*60)
print("Highest BloodPressure recorded",df['BloodPressure'].max())
print("="*60)
print("lowest BloodPressure recorded",df['BloodPressure'].min())
print("="*60)
print("Average HealthScore",df.groupby('Risk')['HealthScore'].mean())
print("="*60)
print(df.groupby('Risk').agg({
  'Age' : "mean",
  'HeartRate':'mean',
  'BloodPressure':'mean',
  'HealthScore':'mean'
  }))
```

# 📌 Key Insights

- Dataset contains 51 patient records.
- Missing HeartRate values were replaced with the average HeartRate.
- Duplicate records were identified and removed.
- High Risk patients have the highest average BloodPressure.
- HealthScore was created using HeartRate and BloodPressure.
- GroupBy and Aggregation were used to summarize patient health statistics.
